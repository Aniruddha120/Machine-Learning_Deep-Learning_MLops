**Line Plot**

Here the Seaborn and matplot has been imported. The dataset has been used from seaborn github datasets.
Dataset links: https://github.com/mwaskom/seaborn-data.git
Seabornfeature used: hue, style, palette, markers
Matplotlib feature used: grid, title

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
y_1 = sns.load_dataset("penguins")
y_1
```

Exemple of matplotlib line graph

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,5,6,7,16]
plt.plot(var,var_1)
plt.show()
```

```python
sns.lineplot(x= "flipper_length_mm",y="bill_length_mm",data=y_1,hue="sex", style="sex", palette="Accent",markers=["*","o"])
plt.grid()
plt.title("Penguine")
plt.show()
```
